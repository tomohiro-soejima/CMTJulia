```julia
using DifferentialEquations
using Plots
```

```julia
function f(u, p, t)
    return p * u
end
```

```julia
u0 = 1.0
t0 = [0.0, 3.0]
p = 0.8
prob = ODEProblem(f, u0, t0, p, abstol=1e-15, reltol=1e-15)
sol = solve(prob)
```

```julia
plot(sol)
println(sol[end])
```
